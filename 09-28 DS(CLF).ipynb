```python
age = dataset['Age']
```

```python
age.mean()
```

```python
class1 = 40 # average of age for class1
```

```python
class2 = 28  # average of age for class2
```

```python
class3 = 24  # average of age for class3
```

```python
def lwmiss(data):
    age = data[0]
    Pclass = data[1]
    if pandas.isnull(age):
        if Pclass == 1:
            return class1
        elif Pclass == 2:
            return class2
        elif Pclass == 3:
            return class3
        
    else:
        return age
```

```python
ap = x[['Age' , 'Pclass']].apply(lwmiss , axis = 1) # here values of age and Pclass are taken as subset 
    # and lwmiss function is applied where age is null according to the class in Pclass
```

```python
x['Age'] = ap
```

```python
xds.dropna() #drop complete row if all values are null
```

```python
from sklearn.model_selection import  train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(xds,y,test_size = 0.33 ,random_state = 42 )
```

```python
from sklearn.linear_model  import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train , y_train)
```

```python
model.coef_
```

```python
# accuracy 
y_pred = model.predict(x_test)
```
