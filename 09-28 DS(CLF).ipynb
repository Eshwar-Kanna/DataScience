```python
class1 = 40
```

```python
class2 = 28
```

```python
class3 = 24
```

```python
def lwmiss(data):
    age = data[0]
    Pclass = data[1]
    if pandas.isnull(age):
        if Pclass == 1:
            return class1
        elif Pclass == 2:
            return class2
        elif Pclass == 3:
            return class3
        
    else:
        return age
```

```python
ap = x[['Age' , 'Pclass']].apply(lwmiss , axis = 1)
```

```python
x['Age'] = ap
```
