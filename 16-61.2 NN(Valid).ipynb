```python
import pandas as pd

dataset = pd.read_csv('wine.csv')

dataset

dataset.columns

y = dataset['Wine']

x = dataset[['Alcohol','Malic.acid','Ash','Acl','Mg','Phenols','Flavanoids','Nonflavanoid.phenols','Proanth','Color.int','Hue','OD','Proline']]

y = pd.get_dummies(y)

from keras.models  import Sequential

model = Sequential()

from keras.layers import Dense

model.get_config()

model.add(
    Dense(
           units = 8,
            activation = 'relu',
            kernel_initializer = 'he_normal',
            input_shape=(13,)
    )
)

model.add(
    Dense(
           units = 6,
            activation = 'relu',
            kernel_initializer = 'he_normal'
    )
)

model.add(
    Dense(
           units = 3,
            activation = 'softmax',
            kernel_initializer = 'he_normal'
    )
)

model.get_config()

model.compile(loss = "categorical_crossentropy" , optimizer = "adam" , metrics = ['accuracy'])

model.fit(x , y , epochs = 54 , validation_split = 0.20)

mloss = model.history.history['loss']

mloss = pd.DataFrame(mloss)

mloss.plot()

```
