```python
import pandas as pd
```

```python
dataset_train = pd.read_csv("Google_Stock_Price_Train.csv" , index_col = 0)
```

```python
dataset_train
```

```python
dataset_test = pd.read_csv("Google_Stock_Price_Test.csv" , index_col = 0)
```

```python
dataset_test
```

```python
training_set = dataset_train[['Open']]
```

```python
testing_set = dataset_test['Open']
```

```python
testing_set
```

```python
training_set.plot()
```

```python
 testing_set.plot()
```

```python
from sklearn.preprocessing import MinMaxScaler 
```

```python
scale = MinMaxScaler(feature_range=(0,1))
```

```python
training_set_scaled = scale.fit_transform(training_set)
```

```python
x_train = []
```

```python
y_train = []
```

```python
for i in range(60 , 1258):
    x_train.append(training_set_scaled[i-60:i,0])
    y_train.append(training_set_scaled[i,0])
```

```python
y_train
```

```python
import numpy as np
```

```python
x_train_array = np.array(x_train)
```

```python
y_train_array = np.array(y_train)
```

```python
x_train_array.shape
```

```python
y_train_array.shape
```

```python
from keras.models import Sequential
```

```python
from keras.layers import LSTM
```

```python
model = Sequential()
```

```python
model.add(LSTM(
                units = 60 ,
                return_sequences = True,
                input_shape = (60, 1)
))
```

```python
model.layers
```

```python
model.summary()
```

```python
model.add(LSTM(
                units = 30,
                return_sequences = True
))
```

```python
model.add(LSTM(units = 30))
```

```python
from keras.layers import Dense
```

```python
model.add(Dense(units = 1))
```

```python
model.layers
```

```python
model.compile(optimizer = 'adam' , loss = 'mean_squared_error')
```

```python
model.fit(x_train_array , y_train_array)
```

```python
model.get_weights
```

```python
model.predict()
```
