```python
from keras.preprocessing import image
```

```python
img = image.load_img("dog.jpeg" , target_size = (224 , 224)) # predicting image should be in the same size and shape
# in similar to the trained images to the model
```

```python
img.size   # image is in 2D It must be in 4D farmat in similar to the images we trained to the model for its prediction
```

```python
type(img)
```

```python
img_np = image.img_to_array(img) # converting image to numpy image
```

```python
img_np.shape  # here the image converted to 3D
```

```python
import numpy as np
```

```python
img_final = np.expand_dims(img_np , axis = 0 ) # this will expand the image by one dimension ie. 2D-3D , 3D-4D , 4d-5D
```

```python
img_final.shape # converted to 4D as per the requirement
```

```python
model.predict(img_final)
```

```python
# we have precreated functions in keras to decode this values and provide the class name for this predicted values
```

```python
from keras.applications.vgg16 import decode_predictions
```

```python
predict1 = model.predict(img_final)
```

```python
decode_predictions(predict1 , top = 3) # this will provide the predicted classes of top 3 accuracy percentages
```

```python
img2 = image.load_img("dog2.jpg" , target_size=(224 , 224))
```

```python
imgnp = image.img_to_array(img2)
```

```python
imgnp.shape
```

```python
image_final = np.expand_dims(imgnp , axis = 0)
```

```python
# in vgg we also have a precreated function for processing the image as required for the vgg model 
```

```python
from keras.applications.vgg16 import preprocess_input
```

```python
image_final = preprocess_input(image_final)
```

```python
predict2 = model.predict(image_final)
```

```python
decode_predictions(predict2 , top=3)
```
